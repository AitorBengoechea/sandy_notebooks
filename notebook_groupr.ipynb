````python
folder = "notebook_groupr_data"
````

````python
filename = "tape91"
file = os.path.join(os.getcwd(), folder, filename)
tape = sandy.Groupr.from_file(file)
````

````python
iwt18, xs18 = tape.get_xs(9437, 18)
xs18.index = xs18.index.left
iwt18.index = iwt18.index.left
````

````python
iwt102, xs102 = tape.get_xs(9437, 102)
xs102.index = xs102.index.left
iwt102.index = iwt102.index.left
````

````python
fig, ax = plt.subplots(figsize=(8, 5))

xs18.plot(drawstyle='steps-pre', color="dodgerblue", ax=ax)
xs102.plot(drawstyle='steps-pre', color="orange", ax=ax)

ax.set_xscale("log")
ax.set_yscale("log")

ax2 = ax.twinx()

iwt18.plot(drawstyle='steps-pre', color="tomato", style='--', alpha=.7, ax=ax2)
ax2.set_yscale("log")
ax2.grid(False)

plt.tight_layout();
````
